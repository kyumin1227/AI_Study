```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3]
y = [1, 2, 3]

x_axis = []
y_axis = []

for w in range(21):
    w = w * 0.1
    result = ((w * x[0] - y[0]) ** 2 + (w * x[1] - y[1]) ** 2 + (w * x[2] - y[2]) ** 2) / 3
    x_axis.append(w)
    y_axis.append(result)

plt.plot(x_axis, y_axis)
plt.show()
```
