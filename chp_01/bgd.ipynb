```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_train = [ np.random.rand() * 10 for _ in range(50)]
y_train = [val + np.random.rand() * 5 for val in x_train]
```

## 데이터 시각화

학습에 사용할 데이터들을 시각화

```python
plt.scatter(x_train, y_train)
plt.show()
```

## 학습

```python
w = 0
learning_late = 0.0001
epoch = 100
w_list = []
```

```python
for _ in range(epoch):
    slope = 0
    for x, y in zip(x_train, y_train):
        slope += x * (w * x - y)
    
    print(slope)
    w -= slope * learning_late
    w_list.append(w)
```

## 기울기 값

epoch에 따른 기울기값의 변화를 확인

```python
plt.plot(list(range(100)), w_list)
plt.show()
```

## 결과값 비교

기울기값으로 그린 직선과 기존의 데이터를 비교

```python
plt.plot(list(range(10)), [w * i for i in range(10)])
plt.scatter(x_train, y_train)
plt.show()
```
