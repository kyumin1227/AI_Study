```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4]
y = [2, 4, 6, 8]

w = 4   # 초기 예측값
loss = 0.0
learning_rate = 0.01    # 학습률
avg = []    # 로스률 저장
epoch = 100
```

```python
for _ in range(epoch):
    slope_sum = 0
    loss = 0
    for x_train, y_train in zip(x, y):
        pred_val = w * x_train

        loss += (pred_val - y_train) ** 2
        
        slope_sum += x_train * (w * x_train - y_train)

    w = w - learning_rate * (slope_sum / len(x))    # 예측 경사도 업데이트

    avg.append(loss / len(x))

print(f"W: {w}")

plt.plot(range(1, epoch + 1), avg)
plt.show()
```
